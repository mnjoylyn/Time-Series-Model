# Time Series Model: Craiglist Dataset

```python
import pandas as pd
```

```python
# Load the Craigslist CSV file into a DataFrame
df = pd.read_csv("craiglist.csv")
```

```python
# Function to fill missing values with median for numerical columns
def fill_missing_numerical(df):
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    for col in numerical_cols:
        median_value = df[col].median()
        df[col].fillna(median_value, inplace=True)
```

```python
# Function to fill missing values with mode for categorical columns
def fill_missing_categorical(df):
    categorical_cols = df.select_dtypes(include=['object']).columns
    for col in categorical_cols:
        mode_value = df[col].mode().iloc[0]
        df[col].fillna(mode_value, inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# Visualize temporal patterns
plt.figure(figsize=(12, 6))
plt.title('Number of Listings Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Listings')
data.resample('D').size().plot(legend=False)  # Resample data to daily frequency
plt.show()
```

```python
# Analyze seasonal trends
plt.figure(figsize=(12, 6))
plt.title('Seasonal Trends')
plt.xlabel('Month')
plt.ylabel('Average Number of Listings')
data.resample('M').size().plot(legend=False)  # Monthly resampling
plt.show()
```

```python
# Analyze demand-supply dynamics by region
plt.figure(figsize=(12, 6))
plt.title('Listings by Region')
plt.xlabel('Region')
plt.ylabel('Number of Listings')
data['region'].value_counts().plot(kind='bar')
plt.xticks(rotation=90)
plt.show()
```

```python
# Analyze demand-supply dynamics by vehicle type
plt.figure(figsize=(12, 6))
plt.title('Listings by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Number of Listings')
data['type'].value_counts().plot(kind='bar')
plt.xticks(rotation=90)
plt.show()
```

```python
# Statistical analysis
# You can calculate various statistics or use libraries like NumPy and SciPy for more in-depth analysis.

# Correlation between numerical variables
numerical_data = data.select_dtypes(include='number')  # Select only numeric columns
corr_matrix = numerical_data.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()

# Statistical summary
summary = numerical_data.describe()
print(summary)
```
